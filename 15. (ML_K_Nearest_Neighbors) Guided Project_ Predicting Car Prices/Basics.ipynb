### Import library and Read csv file:

```python
import pandas as pd
import numpy as np


pd.options.display.max_columns = 99
cols = ['symboling', 'normalized-losses', 'make', 'fuel-type', 'aspiration', 'num-of-doors', 'body-style', 
        'drive-wheels', 'engine-location', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'engine-type', 
        'num-of-cylinders', 'engine-size', 'fuel-system', 'bore', 'stroke', 'compression-rate', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']
cars = pd.read_csv('imports-85.data', names=cols)
cars.head()
```

### Data cleaning:

From the total number of the row, is null value should not be dropped since it is a large portion of entire data frame.
There are 4 null values in price. I will remove these 4 values since we want to predict the price.

```python
cars = cars.dropna(subset=['price'])
cars.isnull().sum()
```

```python
# fill the na data with column mean
cars = cars.fillna(cars.mean())
```

```python
cars.isnull().sum()
```

```python
# Select only the columns with continuous values from - https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.names
continuous_values_cols = ['normalized-losses', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'bore', 'stroke', 'compression-rate', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']
numeric_cars = cars[continuous_values_cols]
```

```python
# Normalize all columnns to range from 0 to 1 except the target column.
price_col = numeric_cars['price']
numeric_cars = (numeric_cars - numeric_cars.min())/(numeric_cars.max() - numeric_cars.min())
numeric_cars['price'] = price_col
```

### Data training:
- k-nearest neighbors

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
```

```python
def knn_train_test(train_col, target_col, df):
    knn = KNeighborsRegressor()
    np.random.seed(1)
        
    # Randomize order of rows in data frame.
    shuffled_index = np.random.permutation(df.index)
    rand_df = df.reindex(shuffled_index)

    # Divide number of rows in half and round.
    last_train_row = int(len(rand_df)/2)
    # Select the first half and set as training set.
    training_set = rand_df.iloc[0:last_train_row]
    # Select the second half and set as test set.
    testing_set = rand_df.iloc[last_train_row:]

    
    # Fit a KNN model using default k value.
    knn.fit(training_set[[train_col]],training_set[target_col])
    # Make predictions using model.
    prediction = knn.predict(testing_set[[train_col]])
    # Calculate and return RMSE.
    mse = mean_squared_error(testing_set[target_col],prediction)
    rmse = np.sqrt(mse)

    return rmse
```

```python
train_cols = numeric_cars.columns.drop('price')
rmse_result = {}
for train_col in train_cols:
    rmse_val = knn_train_test(train_col,'price',numeric_cars)
    rmse_result[train_col] = rmse_val

rmse_results_series = pd.Series(rmse_result)
rmse_results_series.sort_values()
```

### knn_train_test varying k value:

```python
def knn_train_test(train_col, target_col, df):
    
    np.random.seed(1)
        
    # Randomize order of rows in data frame.
    shuffled_index = np.random.permutation(df.index)
    rand_df = df.reindex(shuffled_index)

    # Divide number of rows in half and round.
    last_train_row = int(len(rand_df)/2)
    # Select the first half and set as training set.
    training_set = rand_df.iloc[0:last_train_row]
    # Select the second half and set as test set.
    testing_set = rand_df.iloc[last_train_row:]
    
    # k values:
    ks = [1,3,5,7,9]
    
    rmse_ks = {}
    
    for k in ks:
        knn = KNeighborsRegressor(n_neighbors=k)

        # Fit a KNN model using default k value.
        knn.fit(training_set[[train_col]],training_set[target_col])
        # Make predictions using model.
        prediction = knn.predict(testing_set[[train_col]])
        # Calculate and return RMSE.
        mse = mean_squared_error(testing_set[target_col],prediction)
        rmse = np.sqrt(mse)
        rmse_ks[k] = rmse
        
    return rmse_ks

rmse_ks_result = {}
train_cols = numeric_cars.columns.drop('price')

for train_col in train_cols:
    rmse_val = knn_train_test(train_col,'price',numeric_cars)
    rmse_ks_result[train_col] = rmse_val
    
rmse_ks_result
```

## Visualization:

```python
import matplotlib.pyplot as plt
%matplotlib inline

for k,v in rmse_ks_result.items():
    x = list(v.keys())
    y = list(v.values())
    
    plt.plot(x,y)
    plt.xlabel('k value')
    plt.ylabel('RMSE')
```
